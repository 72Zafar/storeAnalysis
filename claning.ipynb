```python
plt.figure(figsize=(30,9)) 
pro_count = df.groupby(["Product Name"],as_index = False)["Units Sold"].sum().sort_values(by="Units Sold",ascending=False).head(8)
ax = sns.barplot(x="Product Name",y="Units Sold",data=pro_count)
for bar in ax.containers:
    ax.bar_label(bar)
plt.show()
```

```python

r = df.groupby(["Product Category", "Region"], as_index=False)["Units Sold"].sum().sort_values(by="Units Sold", ascending=False).head(8)
plt.figure(figsize=(20,10))
ax = sns.barplot(x="Product Category", y="Units Sold", hue="Region", data=r)
for bar in ax.containers:
    ax.bar_label(bar)
plt.show()

```

```python
T_r = df.groupby(["Product Category"],as_index=False)["Total Revenue"].sum().sort_values(by="Total Revenue",ascending=False).head(8)
plt.figure(figsize=(15,9))
ax = sns.barplot(x="Product Category",y="Total Revenue",data=T_r)
for bar in ax.containers:
    ax.bar_label(bar)
plt.show()    
```

```python
u_prise = df.groupby(["Product Category"],as_index=False)["Unit Price"].sum().sort_values(by="Unit Price",ascending=False).head(8)
plt.figure(figsize=(15,10))
ax = sns.barplot(x="Product Category", y="Unit Price",data=u_prise)
for bar in ax.containers:
    ax.bar_label(bar)
plt.show()    
```

```python

p_method = df.groupby(["Product Category"], as_index=False)["Payment Method"].count().sort_values(by="Payment Method", ascending=False).head(9)
plt.figure(figsize=(20,10))
ax = sns.barplot(x="Product Category", y="Payment Method", data=p_method)
for bar in ax.containers:
    ax.bar_label(bar)
plt.show()
```
